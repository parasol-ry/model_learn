```python
error_xuen_dicts = []
for i in tqdm(data):
    try:
        block = Block.from_dict(i)
        flat_records, ids = BlockHandcraftRecord.from_block(block).to_flat_records(tag_filter2)
        _df = pd.DataFrame(flat_records)

        texts = []
        other_texts = []
        for span_id, other_span_id in ids:
            texts.append(block[int(span_id[0])][int(span_id[1])].text)
            other_texts.append(block[int(other_span_id[0])][int(other_span_id[1])].text)
        text_df = pd.DataFrame({"text": texts, "other_text": other_texts})

        X = _df.iloc[:, :-1].astype(dtypes)
        y = _df.iloc[:, -1]
        probs = model.predict_proba(X)[:, 1]
        preds = (probs > 0.5).astype("int")
        text_df["ids"] = ids
        text_df["pred"] = preds
        text_df["prob"] = probs
        text_df["label"] = y
        diff_df = text_df[text_df["pred"] != text_df["label"]]
        if len(set(diff_df["other_text"])) >= 5:
            error_xuen_dicts.append(i)
    except Exception:
        pass
```
